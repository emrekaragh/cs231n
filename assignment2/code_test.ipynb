```python
# Define the command you want to execute
command = "wget --directory-prefix=drive/MyDrive/Colab\ Notebooks/cs231n/slides https://example-files.online-convert.com/document/txt/example.txt"
command = "wget --directory-prefix=drive/MyDrive/Colab\ Notebooks/cs231n/slides -r -np -nH -R index.html http://cs231n.stanford.edu/slides/"
# command = 'ls -l drive/MyDrive/Colab\ Notebooks/cs231n/slides'

# Run the command and capture the output
result = subprocess.run(command, shell=True, capture_output=True, text=True)

# Print the command output
print("Command Output:")
print(result.stdout)

# Print the command error, if any
if result.stderr:
    print("Command Error:")
    print(result.stderr)
```

```python
import subprocess

while True:
    # Read user input
    command = input("Enter a command (or 'exit' to quit): ")
    
    # Check if the user wants to exit
    if command.lower() == 'exit':
        break
    
    # Run the command and capture the output
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    
    # Print the command output
    print("Command Output:")
    print(result.stdout)
    
    # Print the command error, if any
    if result.stderr:
        print("Command Error:")
        print(result.stderr)
```
